```python
wins_after_batting_first = df[df['winner'] == df['team1']]

wins_count = wins_after_batting_first['team1'].value_counts()

top_3_teams = wins_count.head(3)

plt.figure(figsize=(10, 6))
top_3_teams.plot(kind='bar', color='skyblue')
plt.title('Top 3 Teams with Most Wins After Batting First')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.show()
```

```python
# win % by team


team_wins = df[df['winner'] == df['team1']].groupby('team1').size()
team_matches = df.groupby('team1').size()
win_percentage = (team_wins / team_matches * 100).fillna(0)
print(win_percentage)
```

```python
#Analyze Match Outcomes by Venue

venue_outcome = df.groupby('venue')['result'].value_counts().unstack().fillna(0)
print(venue_outcome)
```

```python
# Evaluate Toss Decision Effectiveness

toss_decision_win = df[df['toss_winner'] == df['team1']].groupby('toss_decision')['winner'].value_counts().unstack().fillna(0)
print(toss_decision_win)
```

```python
# Find Average Target Runs and Overs by Season

avg_target = df.groupby('season')['target_runs'].mean()
avg_overs = df.groupby('season')['target_overs'].mean()
print(avg_target)
print(avg_overs)
```

```python
# Analyze Matches by Match Type

match_type_stats = df['match_type'].value_counts()
print(match_type_stats)
```

```python
# Find the Distribution of Umpires

umpire1_dist = df['umpire1'].value_counts()
umpire2_dist = df['umpire2'].value_counts()
print(umpire1_dist)
print(umpire2_dist)
```

```python
# Identify High-Scoring Matches 

high_scoring_matches = df[df['target_runs'] > 200]
print(high_scoring_matches)
```

```python
# Determine the Most Common Result Margin

result_margin_freq = df['result_margin'].value_counts()
print(result_margin_freq)
```

```python
# Analyze Matches by City

city_match_stats = df['city'].value_counts()
print(city_match_stats)
```

```python
# Evaluate the Frequency of Super Overs

super_over_count = df['super_over'].value_counts()
print(super_over_count)
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='super_over')
plt.title('Frequency of Super Overs')
plt.xlabel
```
